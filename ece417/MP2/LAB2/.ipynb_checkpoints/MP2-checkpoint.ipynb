```python
# Number of hidden states: N
N = 5
# Initialization of parameters: pi, A, mu, Sigma
pi = (1/N) * np.ones((N, 1))
# print(pi)

A = [[0.8, 0.2, 0, 0, 0], [0, 0.8, 0.2, 0, 0],[0, 0, 0.8, 0.2, 0], [0, 0, 0, 0.8, 0.2], [0, 0, 0, 0, 1]]
# print(A)

# Calculate mean and covariance

mu = {}
covariance = {}
one_word = []
cov = 0*np.identity(14)
speaker =['dg','ls','yx','mh']

for i in words:
    for j in type1[0]:
    # for j in speaker:
        for k in num:
            #loop through the files
            file=open("feature/"+j+"/"+j+"_"+i+k+".fea",'r')
            lines=file.readlines()
            #loop through the lines
            for l in range(0,len(lines)):
                new_line=lines[l].split(",")
                for h in range(0,len(new_line)):
                    new_line[h]=float(new_line[h])
                one_word.append(new_line);
    
    aver = np.average(one_word, axis=0);
    mu[i] = aver;
    mu[i] = [np.array(mu[i]) for j in range(N)]
    
    for j in speaker:
        for k in num:
            #loop through the files
            file=open("feature/"+j+"/"+j+"_"+i+k+".fea",'r')
            lines=file.readlines()
            #loop through the lines
            for l in range(0,len(lines)):
                new_line=lines[l].split(",")
                for h in range(0,len(new_line)):
                    new_line[h]=float(new_line[h])
                new_line = new_line - aver;
                cov = cov + np.outer(new_line,new_line)+0.2*np.identity(14);
                #cov = cov + np.cov(new_line,rowvar=False)+0.2*np.identity(14);
    
    cov = cov / len(one_word);
    covariance[i] = cov
    covariance[i] = [np.array(covariance[i]) for j in range(N)];
    one_word = []
    cov = 0*np.identity(14)

```

```python
# N is the number of hidden states
# T is the number of frames in this file

def forward_backward(pi, A, B, N, T, training = True):
    # Each row for 1 state. Columns represent the frames
    alpha = np.zeros((N,T))
    beta = np.zeros((N,T))
    gamma = np.zeros((N,T))
    xi = np.zeros((2*N,T))
    Amat = np.array(A)  # Convert to an np matrix so we can compute inner products
    
#     print(B)
    
    # Initialize alpha
    for i in range(0,N):
        alpha[i,0]=pi[i]*B[i,0]
    # Iterate to calculate all alpha
    for t in range(1,T):
        for i in range(0,N):
            alpha[i,t]=B[i,t]*np.inner(alpha[:,t-1],Amat[:,i])
    
    if training:
        # Initialize beta
        for i in range(0,N):
            beta[i,T-1]=1
        # Iterate to calculate beta
        for t in range(T-2,-1,-1):
            for i in range(0,N):
                beta[i,t]=np.inner(Amat[i,0:N],beta[:,t+1]*B[:,t+1])
        # Calculate gamma
        for t in range(0,T):
            gamma[:,t]=alpha[:,t]*beta[:,t]
            gamma[:,t]=gamma[:,t]/np.sum(gamma[:,t])
        # Calculate xi. Made some changes here
        for t in range(0,T):
            for i in range(0,N):
                for j in range(i,i+2):
                    if t < T-1:
                        if j < N:
                            xi[i+j,t]=alpha[i,t]*Amat[i,j]*B[j,t+1]*beta[j,t+1]
                        else:
                            xi[i+j, t] = 0
                    else:
                        # This means that we are evaluating the last frame. Since q_{t+1} DNE, p = 0 
                        xi[i+j, t] = 0
            xi[:,t]=xi[:,t]/np.sum(xi[:,t])
            
        return [gamma, xi]
    
    else:
        return np.sum(alpha[:, T-1])
            
```

```python
def scaled_forward_backward(pi, A, B, N, T, training = True):
    # Each row for 1 state. Columns represent the frames
    tildealpha=np.zeros((N,T))
    tildebeta=np.zeros((N,T))
    log_g = np.zeros((T))
    baralpha = np.zeros((N,T))
    gamma = np.zeros((N,T))
    xi = np.zeros((2*N,T))
    Amat = np.array(A)  # Convert to an np matrix so we can compute inner products
    
   
    # Initialize alpha
    for i in range(0,N):
        baralpha[i,0]=pi[i]*B[i,0]
    log_g[0] = np.log(np.sum(baralpha[:,0]))
    tildealpha[:,0]=baralpha[:,0]/np.exp(log_g[0])

    # Iterate to calculate all alpha
    for t in range(1,T):
        for i in range(0,N):
            baralpha[i,t]=B[i,t]*np.inner(tildealpha[:,t-1],Amat[:,i])
        log_g[t] = np.log(np.sum(baralpha[:,t]))
        # print("t =", t, "g[t] =", np.exp(log_g[t]))
        tildealpha[:,t]=baralpha[:,t]/np.exp(log_g[t])
    
    if training:
        # Initialize beta
        for i in range(0,N):
            tildebeta[i,T-1] = 1.0
            # tildebeta[i,T-1] = 1/np.exp(log_g[T-1])
        # Iterate to calculate beta
        for t in range(T-2,-1,-1):
            for i in range(0,N):
                tildebeta[i,t]=np.inner(Amat[i,0:N],tildebeta[:,t+1]*B[:,t+1])/np.exp(log_g[t+1])
        # Calculate gamma
        for t in range(0,T):
            gamma[:,t] = tildealpha[:,t]*tildebeta[:,t]
            gamma[:,t] = gamma[:,t]/np.sum(gamma[:,t])
        # Calculate xi. Made some changes here
        for t in range(0,T):
            if t < T-1:
                for i in range(0,N):
                    for j in range(i,i+2):
                        if j < N:
                            xi[i+j,t]=tildealpha[i,t]*Amat[i,j]*B[j,t+1]*tildebeta[j,t+1]
                        else:
                            xi[i+j, t] = 0
                xi[:,t]=xi[:,t]/np.sum(xi[:,t])
            else:
                # This means that we are evaluating the last frame. Since q_{t+1} DNE, p = 0 
                xi[i+j, t] = 0
            
        return [gamma, xi]
    
    else:
#         print("Here")
#         print(log_g[T-1])
        return np.exp(log_g[T-1])
```

```python
# Data Format: X[word][file #]
def likelihood(mu, Sigma, X, N):
    T = len(X)
    B=np.zeros((N,T))
    for t in range(0,T):
        for i in range(0,N):
            B[i,t]=stats.multivariate_normal(mu[i],Sigma[i]).pdf(X[t]) 
    return B
```

```python
# X here is X[word]
# gamma and xi are dictionaries. Each entry for each file
def update_model(gamma, xi, X, N):
    A = np.zeros((N, N))
    mu = [[] for i in range(N)]
    Sigma = [0 for i in range(N)]
    
    # Calculate the denominator in the update formula for each state q = i
    den = {}
    for i in range(0,N):
        den[i] = 0
        for l in gamma.keys():
            den[i] += np.sum(gamma[l][i, :])
        print('i =', i, 'den =', den[i])
    # Update transition matrix
    for i in range(0, N):
        for j in range(i,i+2):
            if j < N: # Made a change here
                for l in gamma.keys():
                    A[i][j] += np.sum(xi[l][i+j, :])
                A[i][j] /= den[i]
    # Update the mean for each state
    for i in range(0,N):
        mu[i] = 0
        for l in gamma.keys():
            mu[i] += np.inner(np.transpose(X[l]),gamma[l][i,:])
        mu[i] /= den[i]
    # Update the covariance matrix for each state
    for i in range(0,N):
        for l in gamma.keys():
            Sigma[i] = 0.2*np.identity(14)
            for t in range(0,len(X[l])):
                Sigma[i] += gamma[l][i,t]*np.outer(X[l][t]-mu[i],X[l][t]-mu[i])
        Sigma[i] /= den[i]
        
    return [A, mu, Sigma]
```

```python
# Data Format: X[word]
def create_HMM(X, pi, A, mu, Sigma, N):
    # Let number of files be L each identified with a key l
    gamma = {}
    xi = {}
    L = len(X)
    # Data Format: X[word][file #]
    for l in range(L):
        # Find out the number of frames in this file: T = X[word][file #].shape[1]
        # T = X[word][file #].shape[1]
        T = len(X[l])
        # X[file #] = X[word][file #]
        B = likelihood(mu, Sigma, X[l], N)
        gamma[l], xi[l] = scaled_forward_backward(pi, A, B, N, T)

    # Data Format: X[word]
    # gamma and xi are dictionaries. Each entry for each file
    A, mu, Sigma = update_model(gamma, xi, X, N)
    
    model = {'A': A, 'mu': mu, 's': Sigma}
    
    return model
```
